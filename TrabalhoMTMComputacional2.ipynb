 **Aplicações Práticas de Sistemas Lineares e Interpolação Polinomial na Previsão do Valor do Dólar**

Iniciamos o codigo implementando as bibliotecas numpy e matplotlib, a o qual serão usadas para realizar os calculos e plotar os pontos resultantes em um gráfico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Aqui, criamos um array chamado dias_novembro que representa os dias do mês de novembro. Cada número no array corresponde a um dia específico.

```python
# COTAÇÃO DO DÓLAR EM 6 DIAS DO MÊS DE NOVEMBRO COM ESTIMATIVA PARA OS PRÓXIMOS 6 DIAS FUTUROS

# Definindo os dias de novembro
dias_novembro = np.array([1, 2, 3, 4, 5, 6])
X = np.linspace(1, 6, num=100)  # Aumentei o número de pontos entre 1 e 6 para uma curva mais suave
cotacao = np.array([[4.9020], [4.8964], [4.8705], [4.9029], [4.9205], [4.8801]])
```

```python
# Construção da matriz A com potência dos dias de novembro
# Cada linha representa um termo polinomial (grau 0, 1, 2, ...)
A_sem_transposicao = np.array([dias_novembro**0, dias_novembro, dias_novembro**2, dias_novembro**3, dias_novembro**4, dias_novembro**5])
A = A_sem_transposicao.T  # Transpondo para que cada coluna represente um termo polinomial

# Coeficientes do polinômio interpolador
coeficientes = np.linalg.solve(A, cotacao)

```


Interpolação polinomial trata de responder a seguinte pergunta:

> **Dados $n+1$ pontos de seu gráfico,**
$$(x_0, y_0),$$ $$(x_1, y_1),$$ $$\vdots,$$ $$(x_n, y_n),$$
**como determinar um polinômio de grau $n$ cujo gráfico passe por esses mesmos pontos?**

<br>

Ou seja, queremos obter coeficientes $a_0, a_1, \ldots, a_n$ tal que o polinômio

$$
P_n(x)= a_0 + a_1x+...+ax^n
$$

satisfaz o sistema de equações

$$
\begin{cases}
P(x_0)=y_0\\
P(x_1)=y_1\\
\quad\vdots\\
P(x_n)=y_n
\end{cases}.
$$

<br>



```python
# Função lambda para o polinômio interpolador
funcao = lambda x: coeficientes[0] + coeficientes[1]*x + coeficientes[2]*x**2 + coeficientes[3]*x**3 + coeficientes[4]*x**4 + coeficientes[5]*x**5
```

```python
# Estimando a cotação para cada dia dos próximos 6 dias considerando uma média
dias_estimativa = np.linspace(6, 9, num=100)
cotacoes_estimadas = [funcao(dia) for dia in dias_estimativa] #aqui eu calculo o valor da funcao para cada um dos pontos nos dias estimados

# Print dos valores estimados para os próximos 6 dias
print("Valores estimados para os próximos 6 dias:")
for dia, cotacao_estimada in zip(dias_estimativa, cotacoes_estimadas):
    print(f"Dia {int(dia)}: {cotacao_estimada}")

# Plotando o gráfico
plt.plot(dias_novembro, cotacao, 'ro', label='Dados Originais')
plt.plot(dias_estimativa, cotacoes_estimadas, label='Estimativas Médias para os Próximos 6 Dias')
plt.plot(X, funcao(X), label='Polinômio Interpolador')
plt.grid()
plt.title('Cotação do Dólar nos Últimos 6 Dias de Novembro e Estimativa para os Próximos 6 Dias (com Média para 12 Dias)')
plt.xlabel('Dia de Novembro')
plt.ylabel('Cotação')
plt.legend()
plt.show()
```

conclusão




o polinomio de 5° grau não teve um desenpenho satisfatorio para 4 dias alem do dados fornecidos, o melhor caso possivel seira utilizar o metodo dos minimos quadrados, pois foi analizado que um polinomio de 5° grau não uma boa escolha para este fim.
